#### Plot both $y=x^2$ and $y=e^x$ in the same plot using normal and logarithmic scale.
**Hint**: Use `set_xscale` and `set_yscale`.

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here

plot_options, (plot_1, plot_2) = plt.subplots(nrows=2, ncols=1, figsize=(12, 4))

month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
fitbit['Months'] = pd.Categorical(fitbit['Months'], categories=month_order, ordered=True)
subcat_steps_per_month = fitbit.groupby('Months')['Steps'].mean().reset_index()
fitbit['Day_Type'] = fitbit['Days'].apply(lambda x: 'Weekend' if x in ['Saturday', 'Sunday'] else 'Weekday')
subcat_act_w_or_we = fitbit.groupby('Day_Type')['Steps'].mean().reset_index()

# We add bar plots in each of the charts

plot_1.bar(subcat_steps_per_month['Months'], subcat_steps_per_month['Steps'])
plot_2.bar(subcat_act_w_or_we['Day_Type'], subcat_act_w_or_we['Steps'])

plt.savefig('../figures/fitbit.png')

plt.show()

# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# your code here

fitbit.columns
```

```python
active_steps = fitbit[['Minutes Lightly Active', 'Minutes Very Active', 'Minutes Sedentary', 'Steps']]
active_steps. head(3)
```

```python
plot_options, (plot) = plt.subplots(nrows=3, ncols=1, figsize=(7, 7))

for i in range(len(active_steps.columns)):
    column = active_steps.columns[i]
    if column != 'Steps':
        plot[i-1].scatter(x=active_steps[column], y=active_steps['Steps'])
        plot[i-1].set_xlabel(column)
        plot[i-1].set_ylabel('Steps')
        plot[i-1].set_title(f'{column} vs Steps')

plt.tight_layout()

plt.show()
```

#### We would like to have in one single plot the summary statistics of the feature `Age`. What kind of plot would you use? Plot it. 

```python
# your code here

sns.boxplot(x='Age', data=numeric_columns)
plt.xlabel('Age')
plt.title('Boxplot of Age')
plt.show()
```

```python
"""
The box contains the central 50% of the data.
Left edge represents the first quantile (Q1), the top the 3rd respectively (Q3)
The line in the middle shows the median (Q2)
Whiskers extend from the box to the minimum and maximum values within a certain range, 
typically defined by 1.5 times the The Interquartile Range (IQR = Q3-Q1). 
Values beyond the whiskers are considered outliers and are plotted individually.
The points beyond the whiskers are the outliers.

"""
```
